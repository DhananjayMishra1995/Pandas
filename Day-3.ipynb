```python
import pandas as pd
```

```python
#We will be using a dummy created data

people = {
    'first_name':['josh','jay','maan'],
    'last_name':['doe','mishra','preet'],
    'email':['joshdoes@email.com','jaymishra@email.com','maanpreet@email.com']
}


df = pd.DataFrame(people)
df
```

```python
#Let us say we want to set the email as our index

df.set_index('email')
```

```python
#But if I run df again, we will not see any changes in the dataframe. 
df
```

```python
#To save the changes we need to use inplace=True

copy_df= df.copy()
#We will see this copy command in later notebooks

copy_df.set_index('email',inplace=True)

copy_df
```

```python
#If we want to check all the indexes of data 
copy_df.index
```

```python
#Setting such index is very useful in cases where we want to get the information about
#one particular record. 
#We will see the difference 

#Let we want to check the details about jaymishra@email.com
#If we are using by-default index we need to remember the index which belong to the user
#which is highly impossible

#in this case it is 1
df.loc[1]
```

```python
#Now we will use copy_df
copy_df.loc['jaymishra@email.com']
```

```python
#In case we want to get our original index 0,1,2, we can use reset_index

copy_df.reset_index(inplace=True)
```
